```python
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
X,y = make_classification(n_features=2,n_samples=200,random_state=42,n_informative=2,n_redundant=0)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
model = GaussianNB()
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```

```python
model.score(X,y)
```

```python
from sklearn.metrics import confusion_matrix,plot_confusion_matrix
```

```python
cnf_matrix = confusion_matrix(y,y_pred)
cnf_matrix
```

```python
plot_confusion_matrix(estimator=model,X=X,y_true=y,normalize='true',cmap=plt.cm.Greens)
```
